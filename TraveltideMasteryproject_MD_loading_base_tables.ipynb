```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming df_session is a DataFrame with the provided columns.
# Note: You may need to clean your data first, for example, converting
# date and time columns to datetime objects for accurate calculations.

# Set a clean style for the plots
sns.set_style("whitegrid")
plt.style.use("ggplot")

# --- Visualizations for df_session ---
print("\n--- Generating Visualizations for df_session ---")

# Figure 1: Distribution plots
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(14, 10))
fig.suptitle('Distribution of Session Metrics', fontsize=16)

# Histogram for page clicks
sns.histplot(df_session['page_clicks'], bins=20, kde=True, ax=axes[0, 0])
axes[0, 0].set_title('Distribution of Page Clicks')
axes[0, 0].set_xlabel('Page Clicks')
axes[0, 0].set_ylabel('Number of Sessions')

# Histogram for nights
sns.histplot(df_session['nights'], bins=10, kde=False, ax=axes[0, 1])
axes[0, 1].set_title('Distribution of Nights Booked')
axes[0, 1].set_xlabel('Nights')
axes[0, 1].set_ylabel('Number of Sessions')

# Histogram for base fare USD
sns.histplot(df_session['base_fare_usd'], bins=20, kde=True, ax=axes[1, 0])
axes[1, 0].set_title('Distribution of Base Fare (USD)')
axes[1, 0].set_xlabel('Base Fare (USD)')
axes[1, 0].set_ylabel('Number of Sessions')

# Histogram for hotel price per room night USD
sns.histplot(df_session['hotel_price_per_room_night_usd'], bins=20, kde=True, ax=axes[1, 1])
axes[1, 1].set_title('Distribution of Hotel Price per Room Night (USD)')
axes[1, 1].set_xlabel('Hotel Price (USD)')
axes[1, 1].set_ylabel('Number of Sessions')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()

# Figure 2: Categorical count plots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 6))
fig.suptitle('Booking and Discount Statuses', fontsize=16)

# Count plot for flight and hotel booked
booked_df = df_session[['flight_booked', 'hotel_booked']].melt(var_name='Booking Type', value_name='Booked')
sns.countplot(x='Booking Type', hue='Booked', data=booked_df, ax=axes[0])
axes[0].set_title('Flight vs. Hotel Booked')
axes[0].set_xlabel('Booking Type')
axes[0].set_ylabel('Count')

# Count plot for flight and hotel discounts
discount_df = df_session[['flight_discount', 'hotel_discount']].melt(var_name='Discount Type', value_name='Applied')
sns.countplot(x='Discount Type', hue='Applied', data=discount_df, ax=axes[1])
axes[1].set_title('Flight vs. Hotel Discount Applied')
axes[1].set_xlabel('Discount Type')
axes[1].set_ylabel('Count')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()

# Figure 3: Relationship plots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 6))
fig.suptitle('Relationships in Session Data', fontsize=16)

# Scatter plot: Page clicks vs. Base fare
sns.scatterplot(x='page_clicks', y='base_fare_usd', data=df_session, ax=axes[0])
axes[0].set_title('Page Clicks vs. Base Fare (USD)')
axes[0].set_xlabel('Page Clicks')
axes[0].set_ylabel('Base Fare (USD)')

# Box plot: Base fare by flight booked
sns.boxplot(x='flight_booked', y='base_fare_usd', data=df_session, ax=axes[1], showfliers=False)
axes[1].set_title('Base Fare by Flight Booked (Outliers Removed)')
axes[1].set_xlabel('Flight Booked')
axes[1].set_ylabel('Base Fare (USD)')


plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Buchungstyp ableiten
df_session["booking_type"] = "Nothing"

df_session.loc[
    (df_session["flight_booked"] == True) & (df_session["hotel_booked"] == False),
    "booking_type"
] = "Only Flight"

df_session.loc[
    (df_session["flight_booked"] == False) & (df_session["hotel_booked"] == True),
    "booking_type"
] = "Only Hotel"

df_session.loc[
    (df_session["flight_booked"] == True) & (df_session["hotel_booked"] == True),
    "booking_type"
] = "Flight & Hotel"

# zählen
booking_counts = df_session["booking_type"].value_counts()
print("Counts:\n", booking_counts)

# optional: Prozent
booking_pct = (booking_counts / booking_counts.sum() * 100).round(2)
print("\nPercent:\n", booking_pct)

# Plot
booking_counts.reindex(["Nothing","Only Flight","Only Hotel","Flight & Hotel"]).plot(kind="bar")
plt.title("Booking Type Distribution")
plt.xlabel("Booking Type")
plt.ylabel("Number of Sessions")
plt.xticks(rotation=0)
plt.show()

# Prozentwerte berechnen
booking_pct = (
    df_session["booking_type"]
    .value_counts(normalize=True)
    .mul(100)
    .round(2)
)

order = ["Nothing", "Only Flight", "Only Hotel", "Flight & Hotel"]
booking_pct = booking_pct.reindex(order)

# Plot
ax = booking_pct.plot(kind="bar")
plt.title("Booking Type Distribution (%)")
plt.xlabel("Booking Type")
plt.ylabel("Percentage")
plt.xticks(rotation=0)

# Prozentlabels auf Balken schreiben
for p in ax.patches:
    height = p.get_height()
    ax.annotate(
        f"{height}%",
        (p.get_x() + p.get_width() / 2, height),
        ha="center",
        va="bottom",
        fontsize=10
    )

plt.show()
```

```python
import pandas as pd
import numpy as np

# --- CONFIG ---
CLICK_CAP = 200          # der vermutete harte Cap
NEAR_CAP = 190           # "nahe am Cap"
SHORT_GAP_SEC = 60       # kurze Luecke zwischen Sessions
FAST_CLICKRATE = 2.0     # Klicks pro Sekunde (anpassen, wenn noetig)
REPEAT_CAP_N = 2         # wie oft Cap pro User, um "botty" zu werden

df = df_session.copy()

# --- 1) Spalten robust finden (TravelTide kann je nach Version anders heissen) ---
def pick_col(candidates):
    for c in candidates:
        if c in df.columns:
            return c
    return None

user_col = pick_col(["user_id", "visitor_id", "customer_id", "uid"])
session_col = pick_col(["session_id", "session", "sid"])
start_col = pick_col(["session_start", "session_start_time", "started_at", "start_time", "timestamp", "ts"])
end_col = pick_col(["session_end", "session_end_time", "ended_at", "end_time"])
clicks_col = pick_col(["page_clicks", "clicks", "num_clicks", "pageviews", "page_views"])

missing = [("user_id", user_col), ("session_id", session_col), ("start", start_col), ("clicks", clicks_col)]
missing = [name for name, col in missing if col is None]
if missing:
    raise ValueError(f"Fehlende Spalten (nicht gefunden): {missing}. "
                     f"Vorhanden sind z.B.: {list(df.columns)[:30]} ...")

# Zeiten sicher in datetime umwandeln (wenn schon datetime, passiert nichts)
df[start_col] = pd.to_datetime(df[start_col], errors="coerce")

if end_col is not None:
    df[end_col] = pd.to_datetime(df[end_col], errors="coerce")

# --- 2) Session-Dauer berechnen (wenn kein session_end da ist: NaN) ---
if end_col is not None:
    df["session_duration_sec"] = (df[end_col] - df[start_col]).dt.total_seconds()
else:
    df["session_duration_sec"] = np.nan

# --- 3) Naechste Session pro User finden (um "neue Session nach Cap" zu erkennen) ---
df = df.sort_values([user_col, start_col])
df["next_session_id"] = df.groupby(user_col)[session_col].shift(-1)
df["next_session_start"] = df.groupby(user_col)[start_col].shift(-1)
df["time_to_next_sec"] = (df["next_session_start"] - df[start_col]).dt.total_seconds()

# --- 4) Flags bauen ---
df["is_capped_200"] = df[clicks_col].eq(CLICK_CAP)
df["near_cap"] = df[clicks_col].ge(NEAR_CAP)

# kurze Folge-Session nach (nahe) Cap
df["short_gap_after_near_cap"] = df["near_cap"] & df["time_to_next_sec"].notna() & (df["time_to_next_sec"] <= SHORT_GAP_SEC)

# Klickrate (nur wenn Dauer vorhanden und >0)
df["click_rate_per_sec"] = np.where(
    (df["session_duration_sec"].notna()) & (df["session_duration_sec"] > 0),
    df[clicks_col] / df["session_duration_sec"],
    np.nan
)
df["too_fast_clickrate"] = df["click_rate_per_sec"].notna() & (df["click_rate_per_sec"] >= FAST_CLICKRATE)

# Wiederholte Cap-Sessions pro User
df["cap_count_user"] = df.groupby(user_col)["is_capped_200"].transform("sum")
df["repeat_caps_user"] = df["cap_count_user"] >= REPEAT_CAP_N

# --- 5)
```
